#### Libs section

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
red_patch = mpatches.Patch(color='red', label='Risky trades')
green_patch = mpatches.Patch(color='green', label='Safe trades')
blue_patch = mpatches.Patch(color='blue', label='price (USD)')

import matplotlib.ticker as mticker


import seaborn as sns
```

#### Loading dataset

```python
data = pd.read_csv("src/data.csv")

#Convert timestamp to datetime format
formattedTime = pd.to_datetime(data["timestamp"],unit='ms').dt.strftime('%B %Y',)
formattedTimeInYears = pd.to_datetime(data["timestamp"],unit='ms').dt.strftime('%Y',)
print(len(formattedTime))
print(len(formattedTimeInYears))
```

#### Graphical analysis

```python

ax = sns.lineplot(x=data["timestamp"], y=data["price"]).set(title='Price of Bitcoin over time (15 minutes)', xlabel='Time', ylabel='Price (USD)', xticklabels=[])
```

```python
ax = plt.plot()
ax = sns.barplot(x=formattedTime, y=data["tradesRisky"], color="red").set(title='Number of risky trades over time (montly)', xlabel='Time', ylabel='risky trades',xticklabels=[])
```

```python
ax = plt.plot()
ax = sns.barplot(x=formattedTime, y=data["tradesSafe"], color="green").set(title='Number of safe trades over time (montly)', xlabel='Time', ylabel='safe trades',xticklabels=[])
```

```python
ax=plt.subplots()
ax=sns.barplot(x=formattedTime, y=data["tradesRisky"], color="red")
ax=sns.barplot(x=formattedTime, y=data["tradesSafe"], color="green")
ax.set(title='Number of risky and safe trades over time (montly)', xlabel='Time', ylabel='trades',xticklabels=[])
ax.legend(handles=[red_patch,green_patch])
```

```python
ax=plt.subplots()
ax=sns.barplot(x=formattedTime, y=data["tradesRisky"], color="red")
ax=sns.barplot(x=formattedTime, y=data["tradesSafe"], color="green")
ax = sns.lineplot(x=formattedTime, y=data["price"])
ax.set(title='Full relationship between price and volume (montly)', xlabel='Time', ylabel='price and trades',xticklabels=[])
ax.legend(handles=[red_patch,green_patch,blue_patch])
```
